```python
class ATM:
    def __init__(self):
        self.balance = 10000
        self.transaction_limit = 3
        self.transactions_done = 0
    
    def display_notes(self, amount):
        notes_500 = amount // 500
        amount = amount % 500
        notes_200 = amount // 200
        amount = amount % 200
        notes_100 = amount // 100
        print("Cash Breakdown:")
        print("‚Çπ500 notes:", notes_500)
        print("‚Çπ200 notes:", notes_200)
        print("‚Çπ100 notes:", notes_100)
    
    def cash_withdrawal(self):
        if self.transactions_done >= self.transaction_limit:
            print("‚ùå Maximum transactions completed for today")
            return
        
        amount = int(input("Enter amount to withdraw: "))
        
        if amount % 100 != 0:
            print("‚ùå Please enter amount in multiples of 100")
        elif amount > 20000:
            print("‚ùå Maximum withdrawal amount is ‚Çπ20,000")
        elif amount > self.balance:
            print("‚ùå Not enough balance in your account")
        else:
            self.balance = self.balance - amount
            self.transactions_done = self.transactions_done + 1
            print("‚úÖ Withdrawal completed successfully")
            self.display_notes(amount)
            print("Current Balance: ‚Çπ", self.balance)
    
    def cash_deposit(self):
        if self.transactions_done >= self.transaction_limit:
            print("‚ùå Maximum transactions completed for today")
            return
        
        amount = int(input("Enter amount to deposit: "))
        
        if amount % 100 != 0:
            print("‚ùå Please enter amount in multiples of 100")
        else:
            self.balance = self.balance + amount
            self.transactions_done = self.transactions_done + 1
            print("‚úÖ Deposit completed successfully")
            self.display_notes(amount)
        
        print("Current Balance: ‚Çπ", self.balance)
    
    def show_menu(self):
        while True:
            print("\n===== ATM SERVICES =====")
            print("1. Deposit Money")
            print("2. Withdraw Money")
            print("3. Check Balance")
            print("0. Exit")
            
            choice = int(input("Select an option: "))
            
            if choice == 1:
                self.cash_deposit()
            elif choice == 2:
                self.cash_withdrawal()
            elif choice == 3:
                print("Current Balance: ‚Çπ", self.balance)
            elif choice == 0:
                print("üôè Thank you for using our service")
                break
            else:
                print("‚ùå Please select a valid option")
            
            if self.transactions_done >= self.transaction_limit:
                print("\n‚ö† You have reached the daily limit of 3 transactions")
                break
            
            user_choice = input("\nWould you like another transaction? (y/n): ").lower()
            if user_choice != 'y':
                print("üôè Thank you for using our service")
                break

# Creating object
atm_machine = ATM()
atm_machine.show_menu()
```
